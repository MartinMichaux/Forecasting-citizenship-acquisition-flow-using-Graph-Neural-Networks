```python
# Rename the country column in df_GDP
df_GDP.rename(columns={'LOCATION': 'country','TIME':'year'}, inplace=True)

# Rename the country column in df_WDI
df_WDI.rename(columns={'Country Name': 'country'}, inplace=True)
```

```python
df_HDI = df_HDI[['iso3', 'country', 'hdicode', 'region',
                 'hdi_2000','hdi_2001','hdi_2002','hdi_2003','hdi_2004','hdi_2005','hdi_2006','hdi_2007','hdi_2008','hdi_2009',
                 'hdi_2010','hdi_2011','hdi_2012','hdi_2013','hdi_2014','hdi_2015','hdi_2016','hdi_2017','hdi_2018','hdi_2019','hdi_2020']]
new_df_HDI = pd.DataFrame(columns=['iso3','feature', 'country', 'hdicode', 'region','year','value'])

for i,row in df_HDI.iterrows():
    for year in range(2000,2021):
        year_str = 'hdi_'+str(year)
        feature = 'human development indication'
        new_row = {'feature':feature,'iso3': row['iso3'], 'country': row['country'], 'hdicode': row['hdicode'], 'region': row['region'], 'year': year,'value':row[year_str]}
        new_row_df = pd.DataFrame([new_row])
        new_df_HDI = pd.concat([new_df_HDI, new_row_df], ignore_index=True)
```

```python
print(new_df_HDI)
```

```python
new_df_WDI = pd.DataFrame(columns=['country', 'Country Code', 'Series Code','Feature','value','year'])

for i,row in df_WDI.iterrows():
    for year in range(2000,2021):
        year_str = str(year)+' [YR'+str(year)+']'
        new_row = {'country': row['country'], 'Country Code': row['Country Code'], 'feature': row['Series'],'Series Code':row['Series Code'], 'year': year,'value':row[year_str]}
        new_row_df = pd.DataFrame([new_row])
        new_df_WDI = pd.concat([new_df_WDI, new_row_df], ignore_index=True)
```

```python
print(new_df_WDI)
```

```python
# Merge df_GDP and df_HDI based on country name
df_merged = pd.merge(df_GDP, new_df_HDI, on="country")
print(df_merged)

# Merge df_merged with df_WDI based on country name
df_final = pd.merge(df_merged, new_df_WDI, on="country")

# Set the country name, time, and features as the index for the final dataframe
df_final.set_index(["country", "year", "feature"], inplace=True)

# Sort the dataframe by the index if desired
df_final.sort_index(inplace=True)
```

```python
df_final.to_csv('output.csv', index=False)
```
